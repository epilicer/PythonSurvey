### Visualizing survey questions

In this section survey questions can be plotted for a various set of dependencies, such as;

* instr: Instructor's identifier; values taken from {1,2,3}
* class: Course code (descriptor); values taken from {1-13} 
* repeat: Number of times the student is taking this course; values taken from {0,1,2,3,...} 
* attendance: Code of the level of attendance; values from {0, 1, 2, 3, 4} 
* difficulty: Level of difficulty of the course as perceived by the student; values taken from {1,2,3,4,5}


```python
plt.figure(figsize=(14, 11))
i = 1
for t in ['instr', 'class', 'nb.repeat', 'attendance', 'difficulty']:
    plt.subplot(3,2,i)
    df[t].plot.hist(alpha=0.7, fc='g', ec='g')
    plt.title(t)
    i+=1
```

```python
# function to plot counts for a question grouped by a dependcy for another dependency
  
def plotCat(xdata, huedata, coldata, headingTitle):
    g = sns.catplot(x=xdata, hue=huedata, col=coldata, 
                    kind='count', palette='ch:.5', #height=4, aspect=1,
                    data=df, legend=False)
    axes = g.axes.flatten()
    #axes[0].set_title("1st repitition")
    #axes[1].set_title("2nd repitition")
    #axes[2].set_title("3rd repitition")
    for ax in axes:
        ax.set_xlabel('evaluation')
    fig = g.fig
    fig.subplots_adjust(top=0.9)
    fig.suptitle(headingTitle, fontsize=14)
    g.add_legend(title=huedata)
    #plt.tight_layout()
    plt.show()
    return g
```

As an example, answers given to the "Question 14" which is grouped by number of times the student is taking this course (nb.repeat) for each instructor (instr)

```python
plotCat('Q14', 'instr', 'nb.repeat', questions[13])
```

### Correlation between survey questions

Generally speaking, correlation between two variables refers to the each variable 'relatedness' in some sense. Correlated variables are those which contain information about each other. The stronger the correlation, the more one variable tells us about the other.

In Pandas, corr function has those of the following methods to find correlation between variables:

- pearson : standard correlation coefficient
- kendall : Kendall Tau correlation coefficient
- spearman : Spearman rank correlation
- callable: callable with input two 1d ndarrays


```python
sns.set(style='white')
#sns.set(style='ticks')

# Compute the correlation matrix
questCols = ['Q'+str(i) for i in range(1,29)]
df_quest = df[questCols]

corr = df_quest.corr()
# Use Spearman instead of default Pearson, if theseare ordinal variables!
#corr = df_quest.corr(method='spearman')

fig, ax = plt.subplots(figsize=(14, 11))
sns.heatmap(corr, cmap='RdBu', annot=True, fmt='.2f')#, vmin=-1, vmax=1)
plt.title("Correlations Between Survey Questions", fontsize=14)
plt.tight_layout()
plt.show()
```

### Clustermap for survey questions

Clustermap method uses a hierarchical clusters to order data by similarity. This reorganizes the data for the rows and columns and displays similar content next to one another for even more depth of understanding the data. Instead of eyeballing the heatmap for which individual benefits are positively or negatively associated, the plot will be segmented into clusters, which is a little easier to look at.

Clustermap includes those of methods; â€˜singleâ€™, â€˜averageâ€™, and â€˜wardâ€™, â€˜wardâ€™ is generally a safe bet when starting out.

```python
sns.clustermap(corr, method='ward', cmap='RdBu', annot=True, figsize=(14,12)) #vmin=-1, vmax=1,
plt.title("Correlations Between Survey Questions", fontsize=14)
plt.show()
```
